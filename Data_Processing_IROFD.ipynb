```python
!pip install yfinance

import yfinance as yf
import pandas as pd

# Download 5 years of daily exchange rate data
eurusd = yf.download('EURUSD=X', start='2020-01-01', end='2025-01-01')
audusd = yf.download('AUDUSD=X', start='2020-01-01', end='2025-01-01')
nzdusd = yf.download('NZDUSD=X', start='2020-01-01', end='2025-01-01')

# Save to CSV for later use
eurusd.to_csv('EURUSD.csv')
audusd.to_csv('AUDUSD.csv')
nzdusd.to_csv('NZDUSD.csv')
```

```python

# Print the data
print("EURUSD Data:")
print(eurusd)
print("\nAUDUSD Data:")
print(audusd)
print("\nNZDUSD Data:")
print(nzdusd)
```

```python
eurusd = eurusd[['Close']].rename(columns={'Close': 'EURUSD'})
audusd = audusd[['Close']].rename(columns={'Close': 'AUDUSD'})
nzdusd = nzdusd[['Close']].rename(columns={'Close': 'NZDUSD'})
```

```python
# Print the data
print("EURUSD Data:")
print(eurusd)
print("\nAUDUSD Data:")
print(audusd)
print("\nNZDUSD Data:")
print(nzdusd)
```

```python

# Merge the DataFrames on the index (Date)
merged_data = pd.merge(eurusd, audusd, left_index=True, right_index=True, how='outer')
merged_data = pd.merge(merged_data, nzdusd, left_index=True, right_index=True, how='outer')

# Print the merged data
print(merged_data)

# Save the merged data to a new CSV file
merged_data.to_csv('merged_exchange_rates.csv')
```

```python
#missing exchange rates -  df is not defined, using merged_data instead
df_ffill = merged_data.ffill()
df_ffill
```

```python
import numpy as np

# Calculate the daily logarithmic returns
daily_returns = df_ffill.pct_change().apply(lambda x: np.log(1 + x))

# Calculate the annualized volatility (standard deviation of daily returns * sqrt(252 trading days))
volatility = daily_returns.std() * np.sqrt(252)

print("\nAnnualized Volatility:")
volatility
```

```python

# ## Volatility Analysis
# Calculate the daily logarithmic returns
daily_returns = df_ffill.pct_change().apply(lambda x: np.log(1 + x))

# Calculate the 30-day rolling volatility
rolling_volatility = daily_returns.rolling(window=30).std() * np.sqrt(252)

print("\n30-Day Rolling Volatility:")
rolling_volatility
```

```python
# Calculate the 3-month (approximately 63 business days) rolling volatility
rolling_volatility_3m = daily_returns.rolling(window=63).std() * np.sqrt(252)

print("\n3-Month Rolling Volatility:")
print(rolling_volatility_3m)

# Calculate the average 3-month volatility over the entire period
average_3m_volatility = rolling_volatility_3m.mean()

print("\nAverage 3-Month Volatility:")
average_3m_volatility
```

```python
# Calculate the 6-month (approximately 126 business days) rolling volatility
rolling_volatility_6m = daily_returns.rolling(window=126).std() * np.sqrt(252)

print("\n6-Month Rolling Volatility:")
print(rolling_volatility_6m)

# Calculate the average 6-month volatility over the entire period
average_6m_volatility = rolling_volatility_6m.mean()

print("\nAverage 6-Month Volatility:")
average_6m_volatility
```

```python

# Function to calculate moving average
def moving_average(data, window):
    return data.rolling(window=window).mean()

# Calculate 30-day moving average
ma_30 = moving_average(df_ffill, 30)
print("\n30-Day Moving Average:\n", ma_30)

# Calculate 90-day moving average
ma_90 = moving_average(df_ffill, 90)
print("\n90-Day Moving Average:\n", ma_90)

# Calculate 180-day moving average
ma_180 = moving_average(df_ffill, 180)
print("\n180-Day Moving Average:\n", ma_180)
```

```python
# Install required packages (uncomment if running fresh)
# !pip install yfinance tensorflow

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from sklearn.prAeprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.metrics import mean_squared_error, mean_absolute_error

# --- Step 1: Load Exchange Rate Data (EURUSD) ---
import yfinance as yf

# Download EUR/USD data (already prepared)
eurusd = yf.download('EURUSD=X', start='2020-01-01', end='2025-01-01')
data = eurusd['Close'].dropna()

# --- Step 2: Train-Test Split ---
train_size = int(len(data) * 0.8)
train, test = data[:train_size], data[train_size:]

# --- Step 3: Fit ARIMA Model manually without auto_arima ---
# (Assuming ARIMA(1,1,1) as an example, change as needed)
arima_order = (1, 1, 1)
arima_fitted = ARIMA(train, order=arima_order).fit()
print(f"Selected ARIMA order: {arima_order}")

train_pred_arima = arima_fitted.fittedvalues

# Forecast on test set
test_forecast_arima = arima_fitted.forecast(steps=len(test))

# --- Step 4: Obtain Residuals ---
train_residuals = train - train_pred_arima

# --- Step 5: Prepare Residuals for ANN ---
def create_supervised(series, lag=5):
    X, y = [], []
    for i in range(lag, len(series)):
        X.append(series[i-lag:i])
        y.append(series[i])
    return np.array(X), np.array(y)

lag = 5
X_train, y_train = create_supervised(train_residuals.values, lag)

# Normalize inputs
scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train)

# --- Step 6: Build ANN Model ---
model = Sequential()
model.add(Dense(10, activation='relu', input_shape=(lag,)))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mse')
model.summary()

# Train ANN
model.fit(X_train_scaled, y_train, epochs=100, batch_size=16, verbose=1)

# --- Step 7: Predict Residuals on Test Set ---
# Prepare inputs
full_series = pd.concat([train_residuals, test - test_forecast_arima])
X_full, _ = create_supervised(full_series.values, lag)
X_test_scaled = scaler.transform(X_full[-len(test):])

residual_forecast_ann = model.predict(X_test_scaled).flatten()

# --- Step 8: Combine ARIMA + ANN Predictions ---
hybrid_forecast = test_forecast_arima.values + residual_forecast_ann

# --- Step 9: Evaluate ---
print("\n--- Evaluation Results ---")
print(f"ARIMA Test MSE: {mean_squared_error(test, test_forecast_arima):.8f}")
print(f"Hybrid Test MSE: {mean_squared_error(test, hybrid_forecast):.8f}")
print(f"ARIMA Test MAE: {mean_absolute_error(test, test_forecast_arima):.8f}")
print(f"Hybrid Test MAE: {mean_absolute_error(test, hybrid_forecast):.8f}")

# --- Step 10: Plot Results ---
plt.figure(figsize=(14,7))
plt.plot(test.index, test.values, label='Actual', marker='o')
plt.plot(test.index, test_forecast_arima, label='ARIMA Forecast', linestyle='--')
plt.plot(test.index, hybrid_forecast, label='Hybrid (ARIMA + ANN)', linestyle='-')
plt.legend()
plt.title('EUR/USD Exchange Rate Forecasting')
plt.xlabel('Date')
plt.ylabel('Exchange Rate')
plt.grid(True)
plt.show()
```
